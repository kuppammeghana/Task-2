Univariate Analysis - Numercial Features

```python
num_cols = ['Salary', '10percentage', '12graduation', '12percentage', 'CollegeID', 'English', 
            'Logical', 'Quant', 'Domain', 'ComputerProgramming', 'ElectronicsAndSemicon', 'conscientiousness', 'agreeableness', 
            'extraversion', 'nueroticism', 'openess_to_experience']

fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(20, 25))
for col, ax in zip(num_cols, axes.flatten()):
    sns.boxplot(x=df[col], ax=ax)
    ax.set_title(col)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(20, 25))
for col, ax in zip(num_cols, axes.flatten()):
    sns.histplot(x=df[col], ax=ax, kde=True)
    ax.set_title(col)
plt.tight_layout()
```

Univariate Analysis - Categorical Features

```python
cat_col = ['Gender', 'JobCity','10board', '12board', 'Degree', 'Specialization', 'CollegeState', 'GraduationYear']

fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(45, 55))
for col, ax in zip(cat_col, axes.flatten()):
    sns.countplot(x=df[col], ax=ax)
    ax.set_title(col)
    ax.tick_params(axis='x', rotation=90)
plt.tight_layout()
```

```python
cse_grad = df[df['Specialization'] == 'computer engineering']
desired_job = ['program analyst trainee', 'software engineere', 'hardware engineer', 'software engineering associate', 'software eng', 'associate software engg', 'associate developer', 'software engg']
cse_job = cse_grad[cse_grad['Designation'].str.lower().isin(desired_job)]
# find the maximum and minimum salaries
max_salary = cse_job['Salary'].max()
min_salary = cse_job['Salary'].min()
print("Maximum salary for CSE graduates in desired job titles: ", max_salary)
print("Minimum salary for CSE graduates in desired job titles: ", min_salary)
```

```python
cse_grads = df[df['Specialization'] == 'computer engineering']
desired_jobs = ['programming analyst', 'software engineer', 'hardware engineer', 'associate engineer']
cse_jobs = cse_grads[cse_grads['Designation'].str.lower().isin(desired_jobs)]
max_salary = cse_jobs['Salary'].max()
min_salary = cse_jobs['Salary'].min()

print("Maximum salary for CSE graduates in desired job titles: ", max_salary)
print("Minimum salary for CSE graduates in desired job titles: ", min_salary)
```

```python
gender_spec = pd.crosstab(df['Gender'], df['Specialization'])
fig, ax = plt.subplots(figsize=(20,25))
gender_spec.plot(kind='bar', stacked=True, rot=0, ax=ax)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Specialization by Gender')
plt.show()
```

```python
from scipy.stats import chi2_contingency
cont_table = pd.crosstab(df['Gender'], df['Specialization'])
# performing chi-square test of independence
chi2, p_val, dof, expected = chi2_contingency(cont_table)
# printing results
print(f"Chi-square statistic: {chi2}")
print(f"p-value: {p_val}")
```

Performing Feature Transformation

```python
data = df
```

```python

from sklearn.preprocessing import StandardScaler
numerical_cols = ['Salary', '10percentage', '12graduation', '12percentage', 'CollegeID', 'collegeGPA', 
                  'GraduationYear', 'English', 'Logical', 'Quant', 'Domain', 'ComputerProgramming', 
                  'ElectronicsAndSemicon', 'ComputerScience', 'MechanicalEngg', 'ElectricalEngg', 
                  'TelecomEngg', 'CivilEngg', 'conscientiousness', 'agreeableness', 'extraversion', 
                  'nueroticism', 'openess_to_experience']
scaler = StandardScaler()
data[numerical_cols] = scaler.fit_transform(data[numerical_cols])
# Create dummy variables for categorical features
categorical_cols = ['Gender', 'Designation', '10board', '12board', 'Degree', 'Specialization', 'CollegeState', 'JobCity', 'DOB']
data = pd.get_dummies(data, columns=categorical_cols)
```
